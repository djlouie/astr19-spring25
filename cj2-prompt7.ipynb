# Coding Journal 2: Prompt 7

Create a Jupyter notebook, import matplotlib. Write cells that 1) create a numpy array x ranging from [0,1] in 100 steps and 2) that defines a function that returns exp(x).  In a new cell use the function to set y=exp(x), and then plot x vs. y. Label the x-axis as “Time [milliseconds]” and the y-axis as “Awesomeness”. Save the figure as a PDF. 

```python
import matplotlib  # Not accessed but doing it for the prompt
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
# define x
x = np.linspace(0,1,100)

# define exp(x)
def exp(x):
    return np.exp(x)
```

```python
# define y
y = exp(x)

# plot the data
plt.plot(x,y)
plt.xlabel("Time [milliseconds]")
plt.ylabel("Awesomeness")

# save figure as PDF
plt.savefig("cj2-prompt7.pdf")

# show figure
plt.show()
```
