Create a Jupyter notebook, import matplotlib. Use numpy to pull 1000 random numbers distributed uniformly between [0,1]. Histogram the random numbers into 100 bins, and plot the histogram. Label your axes and save the figure as a PDF.

```python
# imports
import numpy as np
import matplotlib.pyplot as plt
```

```python
# sample 1000 numbers from a uniform distribution between 0 and 1, with size n
n = 1000
x = np.random.uniform(size=n)
```

```python
# create histogram with bins
width = 0.01  # 1/100 = 0.01 this needs to be the width to get 100 bins
histmin = np.floor(min(x))  # set the min of the lowest value and round down (should be 0)
histmax = np.ceil(max(x))  # set the min of the lowest value and round up (should be 1)

# make the limits of the bins with np.arange (makes a np from histmin -> histmax with steps of width)
bins = np.arange(histmin, histmax + width, width)

# plot the data as a histogram
plt.hist(x, bins=bins, edgecolor="black")
plt.title("Histogram of Uniform Random Samples")
plt.xlabel("x")
plt.ylabel("Number of Samples")
plt.savefig("cj2-prompt9.pdf", bbox_inches="tight")
plt.show()
```

Trying the histogram again with np.linspace instead

```python
# create histogram with bins
num_bins = 101  # number of bins minus one because this defines the boundary numbers
histmin = np.floor(min(x))  # set the min of the lowest value and round down (should be 0)
histmax = np.ceil(max(x))  # set the min of the lowest value and round up (should be 1)

# make the limits of the bins with np.arange (makes a np from histmin -> histmax with steps of width)
bins = np.linspace(histmin, histmax, num_bins)

# plot the data as a histogram
plt.hist(x, bins=bins, edgecolor="yellow")
plt.title("Histogram of Uniform Random Samples")
plt.xlabel("x")
plt.ylabel("Number of Samples")
# plt.savefig("cj2-prompt9.pdf", bbox_inches="tight")
plt.show()
```
