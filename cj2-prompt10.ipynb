# Coding Journal 2 Prompt 10:

Create a Jupyter notebook, import matplotlib. On the numpy website, read about the random number distributions available for pulling random variates.  Select your favorite distribution (other than uniform and normal) and pull 1000 random numbers from that distribution. Histogram the random numbers into 100 bins, and plot the histogram. Label your axes and save the figure as a PDF.

```python
import numpy as np
import matplotlib.pyplot as plt
```

Sample and plot and save a binomial distribution

```python
num_samples = 1000
n = 100 # number of trials
p = 0.5 # probability of coinflips head
x = np.random.binomial(n, p, size=num_samples)
```

```python
hist_min = np.floor(min(x))
hist_max = np.ceil(max(x))
num_bin_boundaries = 101

# define bin boundaries
bin_boundaries = np.linspace(hist_min, hist_max, num_bin_boundaries)

# plot histogram
plt.hist(x, bins=bin_boundaries, edgecolor="black")
plt.title("Binomial Distribution")
plt.xlabel("x")
plt.ylabel("Number of Samples")

# savefig
plt.savefig("cj2-prompt10.pdf", bbox_inches="tight")

plt.show()
```

Sample and plot a poisson distribution

```python
num_samples = 1000
lam = 100 # number of expected events in a given time
x = np.random.poisson(lam=lam, size=num_samples)

hist_min = np.floor(min(x))
hist_max = np.ceil(max(x))
num_bin_boundaries = 101

# define bin boundaries
bin_boundaries = np.linspace(hist_min, hist_max, num_bin_boundaries)

# plot histogram
plt.hist(x, bins=bin_boundaries, edgecolor="black")
plt.title("Poisson Distribution")
plt.xlabel("x")
plt.ylabel("Number of Samples")
plt.show()
```

Sample and plot exponential distribution

```python
num_samples = 1000
beta = 10 # scale, inverse of rate
x = np.random.exponential(scale=beta, size=num_samples)

hist_min = np.floor(min(x))
hist_max = np.ceil(max(x))
num_bin_boundaries = 101

# define bin boundaries
bin_boundaries = np.linspace(hist_min, hist_max, num_bin_boundaries)

# plot histogram
plt.hist(x, bins=bin_boundaries, edgecolor="black")
plt.title("Exponential Distribution")
plt.xlabel("x")
plt.ylabel("Number of Samples")
plt.show()
```
