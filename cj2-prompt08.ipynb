# Coding Journal 2: Prompt 8

Create a Jupyter notebook, import matplotlib. Write cells that 1) create a numpy array x ranging from [0,1] in 100 steps and 2) that defines functions that return sin(x) and cos(x).  In a new cell use matplotlib to create a multipanel plot (1 row, 2 columns), plotting sin(x) vs. x in the left panel and cos(x) vs. x in the right panel. Label the panels with sin(x) and cos(x), and save the figure as a PDF.

```python
import matplotlib  # Not accessed but doing it for the prompt
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

```python
x = np.linspace(0, 1, 100)
```

```python
def sin(x):
    return np.sin(x)

def cos(x):
    return np.cos(x)
```

```python
fig, ax = plt.subplots(1, 2)

# sin(x) vs x in left panel
ax[0].plot(x, sin(x), color="r")
ax[0].set_title("sin(x)")

# cos(x) vs x in the right panel
ax[1].plot(x, cos(x), color="g")
ax[1].set_title("cos(x)")

plt.savefig("cj2-prompt8.pdf", bbox_inches="tight", dpi=300)
plt.show()
```
